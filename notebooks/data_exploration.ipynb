```python
import pandas as pd
import numpy as np
import sys
import os

sys.path.insert(0, os.path.abspath(".."))
from src.const import TEAM_LIST, GROUPPING
```

## Elo change since last world cup

```python
df = pd.read_csv("../data/raw/elo_historical.csv")
df = df[df["team"].isin(TEAM_LIST)]
df = df[df["date"] >= "2022-12-20"]
elo_change = (
    df.groupby("team")["rating"]
    .agg(oldest="first", recent="last")
    .assign(change=lambda d: d["recent"] - d["oldest"])
    .sort_values("change", ascending=False)
)
print(elo_change)
```

```python
print(
    f"Top 5 teams with the biggest Elo rating increase in the past 10 years:\n{elo_change.head()}\n"
)
print(
    f"Bottom 5 teams with the biggest Elo rating decrease in the past 10 years:\n{elo_change.tail()}"
)
```

## Average elo ratings since last world cup

```python
avg_elo = df.groupby("team")["rating"].mean().sort_values(ascending=False)
print(avg_elo)
```

```python
print(
    f"Top 5 teams with the highest average Elo rating since last world cup:\n{avg_elo.head()}\n"
)
print(
    f"Bottom 5 teams with the lowest average Elo rating since last world cup:\n{avg_elo.tail().sort_values()}"
)
```

## Average elo ratings of each group

```python
latest_elo = pd.read_csv("../data/raw/elo_latest.csv")
groups_elo = []

for group, teams in GROUPPING.items():
    group_elo = latest_elo[latest_elo["team"].isin(teams)]
    groups_elo.append(
        {"Group": group, "Teams": teams, "Avg_Elo": np.mean(group_elo["elo_ratings"])}
    )

groups_elo_df = pd.DataFrame(groups_elo)
groups_elo_df = groups_elo_df.sort_values(by="Avg_Elo", ascending=False)
groups_elo_df.to_csv("../data/cleaned/groups_avg_elo.csv", index=False)
print(groups_elo_df)
```
